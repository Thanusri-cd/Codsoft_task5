```python
from IPython import get_ipython
from IPython.display import display

contacts={}
def menu():
    while True:
        print("CONTACT BOOK DETAILS")
        print("1.add a contact")
        print("2.search a contact")
        print("3.update a contact")
        print("4.view a contact")
        print("5.delete a contact")
        print("6.exit")
        choice=input("enter your choice(1/2/3/4)")

        if choice == "1":
            add_contact()
        elif choice== "2":
            search_contact()
        elif choice== "3":
            update_contact()
        elif choice== "4":
            view_contact()
        elif choice== "5":
            delete_contact()
        elif choice== "6":
            print("exiting contact book")
            break
        else:
            print("invalid contact")

def add_contact():
    name=input("enter a name")
    phone=input("enter phone number")
    email=input("enter a email")
    address=input("enter a address")
    contacts[name]={"phone":phone,"email":email,"address":address}
    print("contact added successfully")
def update_contact():
    name=input("enter contact name to update")
    if name in contacts:
        phone=input("enter new phone number")
        email=input("enter a new email")
        address=input("enter new address")
        contacts[name]={"phone":phone,"email":email,"address":address}
        print("contact updated successfully")
    else:
        print("contact not found")

def search_contact():
    name=input("enter name to search")
    if name in contacts:

        print(f"name:{name}")

        for key,value in contacts[name].items():

            print(f"{key}:{value}")
    else:
        print("contact not found")

def  view_contact():
    if not contacts:
        print("no contact found")
        return

    for name,info in contacts.items():

            print(f"name:{name}")
            print(f"phone:{info['phone']}")
            print(f"email:{info['email']}")
            print(f"address:{info['address']}")
            print("-"*20)
def delete_contact():
    name=input("enter the name of the contact to delete")
    if name in contacts:
        del contacts[name]
        print("contact deleted successfully")
    elif name not in contacts:
        print(f"contact '{name}' deleted successfully.")
    else:
        print("contact not found to delete")


menu()
```
